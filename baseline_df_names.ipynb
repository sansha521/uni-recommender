```python
df = pd.read_csv("baseline_df_names.csv")
print(f"Rows before cleaning: {len(df)}")

# Remove rows where name is empty string
df_cleaned = df[df["name"].notna()].reset_index(drop=True)
print(f"Rows after removing empty names: {len(df_cleaned)}")

df_cleaned.head(10)
```

```python
if not os.path.exists("baseline_df_names_cleaned.csv"):
    df_cleaned.to_csv("baseline_df_names_cleaned.csv", index=False, header=False)
```
